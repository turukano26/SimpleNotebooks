```python
!wget -O TaipeiSansTCBeta-Regular.ttf https://drive.google.com/uc?id=1eGAsTN1HBpJAkeVM57_C7ccp7hbgSz3_&export=download

import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import font_manager
from matplotlib.font_manager import fontManager

from google.colab import drive
drive.mount('/content/drive/', force_remount=True)

fontManager.addfont('TaipeiSansTCBeta-Regular.ttf')
mpl.rc('font', family='Taipei Sans TC Beta')
font_manager.findfont('Taipei Sans TC Beta')
```

```python
h = 15
w = 10

for p in range(6):
    fig, ax = plt.subplots(h, w)
    fig.set_size_inches(11.5,18)

    fontP = font_manager.FontProperties()
    fontP.set_family('Taipei Sans TC Beta')
    fontP.set_size(32)

    for i in range(h):
        for j in range(w):
            ax[i,j].text(s=text[w*i+j+w*h*p], y=0.2, x=0.2, fontproperties=fontP)
            ax[i,j].axis('off')

    plt.show()
    #plt.savefig("drive/MyDrive/ChineseCharPosters/" + str(p) + ".png")
```
